# Annotation

Research questions: What elements are commonly found in TikTok food review videos that went viral in 2023?

Annotations:

Calculate descriptive statistics for each sub-component.

Determine the percentage of appearances for each sub-component.

Camera Shots Analysis: Identify the types of camera shots used and their frequency. Visualize this information.

Posting Time Analysis: Create histograms and box plots to analyze the distribution of posting times.

Hashtag Analysis: Identify the most frequently used words in hashtags and visualize them using a word cloud.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up clean style
sns.set(style="whitegrid")
```

```python
# Load the annotation Excel file
df = pd.read_excel("Annotation.xlsx")

# Use the first row as column headers
df.columns = df.iloc[0]
data = df[1:]
data.reset_index(drop=True, inplace=True)
```

```python
data_clean = data.head(129)
data_clean
```

```python
components_col = data_clean.iloc[:, 4:-2]
```

```python
components_col.head()
```

```python
# Print the column names to find the correct index for the audio column
print(components_col.columns)
```

```python
column_resolution = components_col.columns[15]
column_graphic = components_col.columns[16]
column_audio = components_col.columns[17]


# Create a list of column names
categorical_col = [column_resolution, column_graphic, column_audio]

```

```python
categorical_col = [
        'Resolution',
        'Type of camera shots (close-ups, wide shots, aerial views)',
        'On-screen graphics (like memes or emojis)',
        'Presence of Background Music'
    ]
```

```python
categorical_col
```

```python
# Drop the columns
binary_component_col = components_col.drop(categorical_col, axis=1)
```

```python
binary_component_col
```

```python
counts = binary_component_col.sum(axis=0)
most_common_components = counts.sort_values(ascending=False)
```

```python
most_common_components
```

```python
plt.figure(figsize=(10, 5))  # Set the figure size
most_common_components.plot(kind='bar')  # Plot all components
plt.title('Most Common Components in Viral Videos')
plt.xlabel('Components')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Calculate the percentage
percentage = (most_common_components / len(binary_component_col)) * 100

# Print the result
print(percentage)
```

```python
# Set the figure size
plt.figure(figsize=(10, 5))

# Plot all components
percentage.plot(kind='bar')

# Set the title and labels
plt.title('Percentage of Videos Each Component Appears In')
plt.xlabel('Components')
plt.ylabel('Percentage')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot the frequencies
most_common_components.plot(kind='bar', color='blue', ax=ax1)
ax1.set_ylabel('Frequency', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis
ax2 = ax1.twinx()

# Plot the percentages
percentage.plot(kind='bar', color='red', ax=ax2, alpha=0.3)
ax2.set_ylabel('Percentage', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Set the title and labels
plt.title('Frequency and Percentage of Each Component in Viral Videos')
plt.xlabel('Components')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
# Define transcript-related columns
transcript_cols = [
    "Appearance of name of food/cuisine",
    "Appearance of adjectives describing food",
    "Appearance of brand names",
    "Mentions of locations",
    "Mentions of prices",
    "Mention of ingredients or cách chế biến"
]

# Use the correct DataFrame (binary_component_col) to select the transcript columns
# Convert values to numeric for summing - it seems these columns were already intended to be numeric (0/1) based on previous uses of binary_component_col
# So we can directly sum. If there were other values, apply(pd.to_numeric, errors='coerce') would be necessary.
transcript_counts = binary_component_col[transcript_cols].sum()

# Plot a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=transcript_counts.index, y=transcript_counts.values)
plt.title("Frequency of Transcript Components in TikTok Food Review Videos")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right') # Adjust rotation and horizontal alignment
plt.tight_layout()
plt.show()
```

```python
# Count unique values for overall audio clarity
# Use the data_clean DataFrame which should contain the 'Overall Audio Clarity' column
# Check if the column exists before accessing it, or just assume it does based on the structure
try:
    audio_clarity_counts = data_clean["Overall Audio Clarity"].value_counts()

    # Plot a pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(audio_clarity_counts, labels=audio_clarity_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title("Distribution of Overall Audio Clarity Ratings")
    plt.axis('equal')
    plt.show()

except KeyError:
    print("Error: 'Overall Audio Clarity' column not found in data_clean DataFrame.")
    # Optional: Print column names to debug
    # print(data_clean.columns)
```

```python
# Calculate the sum for each component
sum_trans = trans_comp.sum(axis=0)
sum_audio_trans = audiotrans_comp.sum(axis=0)
sum_visual = visual_comp.sum(axis=0)
sum_audio = audio_comp.sum(axis=0)

# Create a DataFrame for the sums
df = pd.DataFrame({'Text-Transcript': sum_trans, 'Audio-Transcript': sum_audio_trans, 'Visual': sum_visual, 'Audio': sum_audio})

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(15, 7))

# Plot a stacked bar chart
df.plot(kind='bar', stacked=True, ax=ax)

# Set the title and labels
plt.title('Comparison of Components in Viral Videos')
plt.xlabel('Components')
plt.ylabel('Frequency')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

Methods:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_excel('/content/Survey-3.xlsx')
data
```

```python
cols_before_age = data.columns.tolist()[:data.columns.get_loc('Your age')]
data = data.drop(cols_before_age, axis=1)
data
```

```python
Trans_components = data[data.columns[6:35]]
Trans_components
```

```python
descriptive_stats_trans = Trans_components.describe()
descriptive_stats_trans
```

```python
import matplotlib.pyplot as plt

# Visualizing with box plots
plt.figure(figsize=(12, 8))
Trans_components.boxplot()
plt.xticks(rotation=90)
plt.ylim(1, 10)  # Setting y-axis limits to show the range from 1 to 10
plt.title('Box Plot of Each Factor')
plt.show()
```
